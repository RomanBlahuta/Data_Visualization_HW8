```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('../data/1.4_02.csv')
df.columns = ['Region', 'Year', 'Agriculture', 'Industry', 'Building', 'Services', 'Governing']

df.head()
```

```python
input_radio = alt.binding_radio(options = df.Year.unique(), name='Select year:')
select_year = alt.selection_single(name="Year", fields = ['Year'], bind=input_radio, init = {'Year': df.Year.min()})
```

```python
chart = alt.Chart(
    df,
).transform_fold(
    ['Agriculture', 'Industry', 'Building', 'Services', 'Governing'],
).mark_bar().encode(
    x = alt.X(
        field = 'Region',
        type = 'nominal',
        axis=alt.Axis(
            labelFontSize=14,
        )
    ),
    y = alt.Y(
        field='value',
        type='quantitative',
        stack = 'normalize',
        title='GVA',
        axis=alt.Axis(
            labelExpr="datum.value * 100 + '%'",
            labelFontSize=14,
        )
    ),
    color = alt.Color(
        field = 'key',
        type = 'nominal',
        title='GVA Components',
        scale=alt.Scale(
            scheme='spectral'
        )
    ),
    tooltip = alt.Tooltip(
        field = 'value', 
        type = 'quantitative',
    ),
).add_selection(
    select_year,
).transform_filter(
    select_year,
).properties(
    height=500,
    padding=48,
    title='Change in the structure of Ukraine\'s GVA by region in 2017 compared to 2010'
).configure_title(
    fontSize=24,
    dy=-30,
).configure_legend(
    labelFontSize=14,
    titleFontSize=16,
)

chart.configure_view(
    strokeWidth=0
)
```
